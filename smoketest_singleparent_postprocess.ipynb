```python
qids = []
lfs = []
with open('../grailqa_dev/smoketest/bert_vp_smoketest_extended_singleparent.txt', 'r') as file:
    for line in file:
        result = json.loads(line)
        qids.append(result['qid'])
        lfs.append(result['logical_form'])
        
resultsdf = pd.DataFrame({'qid':qids, 'logical_form':lfs})
```

```python
sparql_list = []

for i, row in resultsdf.iterrows():
    lisp = row.logical_form
    try:
        sparql = logic.lisp_to_sparql(lisp)
    except:
        print('sparql conversion failed')
        print(lisp)
        sparql = 'None'
    sparql_list.append(sparql)
```

```python
resultsdf['sparql'] = sparql_list
```

```python
for i, row in resultsdf.iterrows():
    print('#######', row.qid, '#########################')
    print('Question: ', questions[row.qid])
    print('logical form: \n', row.logical_form)
    print('SPARQL: \n', row.sparql)
```
